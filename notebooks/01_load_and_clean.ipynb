Load the data in chunks, due to many data. Filtered by region. (UK)

```python
import pandas as pd

file_path = "../data/raw/charts.csv"

chunks = []
# process the file in chunks to avoid memory issues reading 500'000 rows at a time
for chunk in pd.read_csv(file_path, chunksize=500_000):
    
    # keep only UK rows
    chunk = chunk[chunk["region"] == "gb"]

    # save the filtered chunks
    chunks.append(chunk)

# combine all chunks into a single DataFrame
uk = pd.concat(chunks, ignore_index=True)

# show first rows of the resulting DataFrame
uk.head()
```

Create the lable (Top 100 vs Rest)

```python
# Create the ML label: 1 = Top100, 0 = rank > 100 .asType(int) converts boolean to integer
uk["is_top100"] = (uk["rank"] <= 100).astype(int)

uk[["rank", "is_top100"]].head()
```
